```python
import pandas as pd
import os
datapath = "../data/ml-1m/"


head = ["UserID", "Gender", "Age", "Occupation", "Zip-code"]

users = pd.read_csv(os.path.join(datapath, "users.dat"), sep="::", engine="python", names=head, usecols=[0, 1, 2, 3])
users
```

```python
pd.get_dummies(users, columns=["Age", "Gender", "Occupation"], dummy_na=False)
```

```python
# 稀疏类别特征编码

cat_cols = ["Gender", "Age", "Occupation"]




mappings = {
    col: dict(zip(values, range(len(values))))
    for col, values in map(lambda col: (col, users[col].unique()), cat_cols)
}

for col, mapping in mappings.items():
    users[col] = users[col].map(mapping.get)
users

```

```python
# 统计每一列独有的值
emb_counts = [len(users[c].unique()) for c in cat_cols]
# 与下述方法一致
# emb_counts = [len(mappings[c]) for c in cat_cols]
emb_counts
```
